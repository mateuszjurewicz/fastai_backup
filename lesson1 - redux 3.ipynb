```python
import csv
import numpy as np
```

```python
# import utils
from utils import *
# Import and instantiate vgg instance
from vgg16 import Vgg16

vgg = Vgg16()
```

```python
batch_size = 64
nr_epochs = 2

train_path = "/home/ubuntu/kaggle/train"
val_path = "/home/ubuntu/kaggle/valid"

# get batches of images to train on
t_batches = vgg.get_batches(train_path, batch_size=batch_size)
v_batches = vgg.get_batches(val_path, batch_size=batch_size)
```

```python
# fine tune the vgg to our classes
vgg.finetune(t_batches)
```

```python
# we can set the lr manually
vgg.model.optimizer.lr = 0.01
```

#### Train the vgg

```python
# we don't want to retrain our model all the time, so let's save the weights
last_weights_file = None
model_path = "/home/ubuntu/kaggle/models/"

# train the model times nr of epochs
for epoch in range(nr_epochs):
    print "Epoch nr %d" % epoch
    vgg.fit(t_batches, v_batches, nb_epoch=1)
    last_weights_file = "weights_%d.h5" % epoch
    vgg.model.save_weights(model_path + last_weights_file)
    
print "Completed %d epoch fits" %nr_epochs
```

#### Alternatively load up previous weights

```python
# weights
weights_file = "/home/ubuntu/kaggle/models/weights_1.h5"
vgg.model.load_weights(weights_file)
```

```python
# use the trained model to make predictions, using the function that Jeremy Howard added to the vgg script
test_path = "/home/ubuntu/kaggle/test"
batches, preds = vgg.test(test_path, batch_size=batch_size)
```

```python
# the vgg.test function returns filenames in batches.filenames and probabilities in the second returned value
# the probabilities are ordered by how we structured the folders (we can check which is 0 idx and which is 1 idx by looking
from PIL import Image

print(preds[0:5])
print(batches.filenames[0:5])
```

```python
print("wow")
```

```python
batches.filenames
filenames = batches.filenames
```

```python
# Strip the image ids appropriately (remove unknown/ and .jpg)
image_ids_jpg = [filename[8:] for filename in filenames]
image_ids = [int(filename[:-4]) for filename in image_ids_jpg]
print(image_ids)
```

```python
# grab the probability that the image is a dog (it's the  second number in each pair)
# use PIL to confirm that assumption (you can see in preds what certainty the model had, they are ordered)
print(preds[1][1])
Image.open(test_path + "/" + filenames[1])
```

```python
# actually grab the  probabilities into a list
dog_prob = [example_pred[1] for example_pred in preds]
```

```python
dog_prob = [round(example_prob, 3) for example_prob in dog_prob]
print(dog_prob)
```

```python
# sanity checks
print(len(dog_prob))
print(len(image_ids))
```

```python
# write the results to a csv file
submission_filename = "test_submission.csv"
columns = ["id", "label"]
with open(submission_filename, "w") as f:
    writer = csv.DictWriter(f, fieldnames=columns)
    
    # write header
    writer.writeheader()
    
    # write each line
    for n in range(len(dog_prob)):
        writer.writerow({"id": image_ids[n], "label": dog_prob[n]})
```

```python
# we can also use more numpy trickery
submissions_file_2 = "test_submission_2.csv"

# numpy has a coo function that lets us concatenate two lists into one list of 2-elem lists, and we can specify if we want
# it to be rows or columns (axis 0 or 1)
stacked_submissions = np.stack([image_ids, dog_prob], axis=1)
stacked_submissions
```

```python
# numpy can also save directly to a csv file
submission_file_name = submissions_file_2
np.savetxt(submission_file_name, stacked_submissions, fmt="%d,%.5f", header="id,label", comments="")
```
