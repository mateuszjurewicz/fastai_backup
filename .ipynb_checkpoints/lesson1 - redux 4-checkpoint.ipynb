## Train on sample
Use the sample data to finetune vgg, save weights.

```python
# set parameters
batch_size = 64
nr_epochs = 2
```

## Train on train set
Use the actual provided train set to train vgg.

```python
# set parameters for training on train set
batch_size_train = 64
nr_epochs_train = 5
```

```python
path_to_train = path_to_data + "train/"
path_to_valid = path_to_data + "valid/"
```

```python
# instantiate another vgg
vgg = Vgg16()

# get batches
train_batches = vgg.get_batches(path_to_train, batch_size=batch_size_train)
valid_batches = vgg.get_batches(path_to_valid, batch_size=batch_size_train)
```

```python
# finetune to our 2 classes
vgg.finetune(train_batches)
```

```python
# let's add a learning rate
vgg.model.optimizer.lr = 0.01
```

```python
# train the model on main data
vgg.fit(train_batches, valid_batches, nb_epoch=nr_epochs_train)
```

```python
# save weights
vgg.model.save_weights(path_to_data + "/results/ft2.h5")
```

## Predict on validation set
Let's load our best weights and make predictions on the validation set.

```python
best_weights_path = None
# best_weights_path = path_to_data + "/results/ft1.h5"
```

```python
# load the model with the best weights
vgg.model.load_weights(best_weights_path)
```

#### Obtain our predictions and actual classes

```python
# predict on the validation set (which is grouped into classes)
val_batches, probs = vgg.test(path_to_valid, batch_size=batch_size_train)
```

```python
# we can't directly print elements of val_batches apparently...
filenames = val_batches.filenames
print(filenames[:5])

# but we can see the probabilities  (which are the thing we'll be using for our submission)
print(probs[0:5])
```

```python
# we can also get the actual classes from our batches!
actual_classes = val_batches.classes

# you can see that cats were first, and dogs were last
print(actual_classes[:10])
print(actual_classes[-10:])
```

#### Cool column slicing

```python
# now let's turn the probabilities into either 1 or 0

# first we can extract (for all examples, hence [:]) the first (zeroth) column, by specifying its index after comma
our_predictions = probs[:,0]
our_labels = np.round(1-our_predictions)

print(our_predictions[:5])
# notice that we had to invert the actual number by subtracting our prediction from 1. That's just because we dropped
# the column with dog confidence. So we were left with cat and the way the labels work is tha cat was 0 and dog 1
```

#### F1 score on valid

```python
# obtain f1 score on validation set
print("F1 on valid: {}".format(f1_score(actual_classes, our_labels)))
```

## Investigate predictions
by looking at some of the images that we classified correctly and incorrectly with both high and low confidence.

```python
# grab things fromp reprocessingras' preprocessing that help with images
from keras.preprocessing import image
%matplotlib inline

#Helper function to plot images by index in the validation set 
#Plots is a helper function in utils.py
def plots_idx(indices, titles=None):
    utils.plots([image.load_img(path_to_valid + filenames[i]) for i in indices], titles=titles)
    
# nr of images to view in each visualisation
n_view = 4
```

#### Investigate correct predictions

```python
# grab all indexes our_predictionsictionsictions we've correctly predicted the class
correct = np.where(our_labels==actual_classes)[0]

# show how many we got right (redundant with the F1 score)
print("Found {} correctly classified images".format(len(correct)))

# grab random n correct indices
indexes = np.random.permutation(correct)[:n_view]

# plot them (show the image and the label)
plots_idx(indexes, our_predictions[indexes])
```

#### Investigate incorrect predictions

```python
# grab all the indices where we predicted incorrectly
incorrect = np.where(our_labels!=actual_classes)[0]

# show how many we got wrong#$
print("Found {} incorrectly classified images".format(len(incorrect)))

# grab random n indices
indexes = np.random.permutation(incorrect)[:n_view]

# plot them
plots_idx(indexes, our_predictions[indexes])
```

#### Investigate high confidence correct dogs
It's not bullshit, I just mixed up our_predictions and our_labels!

```python
# grab all the indices where we correctly classified dogs
dog_correct = np.where((our_labels==1) & (our_labels==actual_classes))[0]

# sort the indices to get highest confidence (we are looking for the lowest values, unlike with cats)
most_confident_dog_correct = np.argsort(our_predictions[dog_correct])[:n_view]

# plot
plots_idx(dog_correct[most_confident_dog_correct], our_predictions[dog_correct][most_confident_dog_correct])
```

#### Investigate high confidence correct cats
Not bullshit, I mixed up our_labels and our_predictions

```python
# grab the indices of correct cat predictions
correct_cats = np.where((our_labels==0) & (our_labels==actual_classes))[0]

# sort them
most_correct_cats = np.argsort(our_predictions[correct_cats])[::-1][:n_view]

# plot
plots_idx(correct_cats[most_correct_cats], our_predictions[correct_cats][most_correct_cats])
```

#### Most uncertain

```python
most_uncertain = np.argsort(np.abs(our_predictions-0.5))
plots_idx(most_uncertain[:n_view], our_predictions[most_uncertain])
```

#### Confident but incorrectly labeled as cats

```python
wrong_cats = np.where((our_labels==0) & (our_labels!=actual_classes))[0]
print("Found {} incorrectly labeled as cats".format(len(wrong_cats)))
conf_wrong_cats = np.argsort(our_predictions[wrong_cats])[:n_view]
plots_idx(wrong_cats[conf_wrong_cats], our_predictions[wrong_cats][conf_wrong_cats])

# there are some incorrectly labeled images in our dataset, methinks
```

#### Confident but incorrectly labeled as dogs

```python
wrong_as_dogs = np.where((our_labels==1) & (our_labels!=actual_classes))[0]
print("Found {} incorrectly labeled as dogs".format(len(wrong_as_dogs)))

confident_wrong_dogs = np.argsort(our_predictions[wrong_as_dogs])[:n_view]
plots_idx(wrong_as_dogs[confident_wrong_dogs], our_predictions[wrong_as_dogs][confident_wrong_dogs])
```

## Submit to Kaggle
We will want to do some tricks because it's log loss.

```python
# they want the id of the image and the probability that it's a dog
# dog probs are the second column (1st index)
is_dog = test_probs[:,1]
print(test_probs)
print(is_dog)
```

```python
# let's visualize some of our is_dog probabilities
print "Raw Predictions: " + str(is_dog[:5])
print "Mid Predictions: " + str(is_dog[(is_dog < .6) & (is_dog > .4)])
print "Edge Predictions: " + str(is_dog[(is_dog == 1) | (is_dog == 0)])
```

```python
# We have a lot of very certain predictions. Log loss will punish us if we get them wrong.
# so let's clip the top probabilities a bit.
is_dog = is_dog.clip(min=0.02, max=0.98)

# clip means that all values in the array that are below min become the value of min and same with max
print "Raw Predictions: " + str(is_dog[:5])
print "Mid Predictions: " + str(is_dog[(is_dog < .6) & (is_dog > .4)])
print "Edge Predictions: " + str(is_dog[(is_dog == 1) | (is_dog == 0)])
print(is_dog)
```

```python
# extract the file names from the paths:
ids = np.array([int(f[8:f.find('.')]) for f in test_filenames])
print(ids)
```

```python
# join the ids and the probs that is dog
subm = np.stack([ids, is_dog], axis=1)
subm[:10]
```

```python
# save the file
submission_file_name = 'submission5.csv'
np.savetxt(path_to_data + "results/" + submission_file_name, subm, fmt='%d,%.5f', header='id,label', comments='')
```
